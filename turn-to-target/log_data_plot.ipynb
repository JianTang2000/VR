```python
with open(r"C:\Users\jiant\Desktop\data\turn-to-target\10-17-continuous verbal.txt", 'r', encoding='UTF-8') as f:
    logs = f.read().split('\n')
    logs_non_blank_line = [i for i in logs if i != '']
print("start processing file is : 10-17-continuous verbal.txt")
tmp_list = [i for i in logs_non_blank_line if i.__contains__("************ CoinSpin.cs, turn-in-space exp log : Coin init at random angle:")]
print(f"turn to target exp num is : {len(tmp_list)}")
print(f"non-blank line count is : {len(logs_non_blank_line)}")

list_target_angle = [None for i in tmp_list]  # 数组长度为实验次数
list_time = [None for i in tmp_list]  # 数组长度为实验次数
list_yaw = [None for i in tmp_list]  # 数组长度为实验次数

time_i = []  # 数组长度为 具体某次转向实验 的数据长度，每次都可能会不一样
yaw_i = []  # 数组长度为 具体某次转向实验 的数据长度，每次都可能会不一样

current_test_index = -1
for i in range(len(logs_non_blank_line)):
    if logs_non_blank_line[i] == "=======this is a gap=======":
        # 存起来上一次的数据
        if current_test_index != -1:
            list_time[current_test_index] = time_i
            list_yaw[current_test_index] = yaw_i
        # index 增加，并清空临时数组放下一轮的数据
        current_test_index += 1
        time_i = []
        yaw_i = []
        print(f"processing {current_test_index} th test data... ")
        continue
    if logs_non_blank_line[i].__contains__("************ CoinSpin.cs, turn-in-space exp log : Coin init at random angle:"):
        list_target_angle[current_test_index] = logs_non_blank_line[i].split("random angle:")[1]
        continue
    if logs_non_blank_line[i].__contains__("current yaw from IMU is ......."):
        part_2 = logs_non_blank_line[i].split(".......")[1]
        this_timestamp = int(part_2.split("<=====>")[0])
        this_yaw = float(part_2.split("<=====>")[1])
        time_i.append(this_timestamp)
        yaw_i.append(this_yaw)
    if i == len(logs_non_blank_line) - 1:
        print("all lines checked! save data...")
        list_time[current_test_index] = time_i
        list_yaw[current_test_index] = yaw_i
        current_test_index += 1
        time_i = []
        yaw_i = []
        print("save data finished!")

print("file 10-17-continuous verbal.txt process finished!")
print(f"random target seq is {list_target_angle}")

continuous_verbal_list_target_angle = list_target_angle
continuous_verbal_list_time = list_time
continuous_verbal_list_yaw = list_yaw

# #####################
with open(r"C:\Users\jiant\Desktop\data\turn-to-target\10-17-continuous cu.txt", 'r', encoding='UTF-8') as f:
    logs = f.read().split('\n')
    logs_non_blank_line = [i for i in logs if i != '']
print("start processing file is : 10-17-continuous cu.txt")
tmp_list = [i for i in logs_non_blank_line if i.__contains__("************ CoinSpin.cs, turn-in-space exp log : Coin init at random angle:")]
print(f"turn to target exp num is : {len(tmp_list)}")
print(f"non-blank line count is : {len(logs_non_blank_line)}")

list_target_angle = [None for i in tmp_list]  # 数组长度为实验次数
list_time = [None for i in tmp_list]  # 数组长度为实验次数
list_yaw = [None for i in tmp_list]  # 数组长度为实验次数

time_i = []  # 数组长度为 具体某次转向实验 的数据长度，每次都可能会不一样
yaw_i = []  # 数组长度为 具体某次转向实验 的数据长度，每次都可能会不一样

current_test_index = -1
for i in range(len(logs_non_blank_line)):
    if logs_non_blank_line[i] == "=======this is a gap=======":
        # 存起来上一次的数据
        if current_test_index != -1:
            list_time[current_test_index] = time_i
            list_yaw[current_test_index] = yaw_i
        # index 增加，并清空临时数组放下一轮的数据
        current_test_index += 1
        time_i = []
        yaw_i = []
        print(f"processing {current_test_index} th test data... ")
        continue
    if logs_non_blank_line[i].__contains__("************ CoinSpin.cs, turn-in-space exp log : Coin init at random angle:"):
        list_target_angle[current_test_index] = logs_non_blank_line[i].split("random angle:")[1]
        continue
    if logs_non_blank_line[i].__contains__("current yaw from IMU is ......."):
        part_2 = logs_non_blank_line[i].split(".......")[1]
        this_timestamp = int(part_2.split("<=====>")[0])
        this_yaw = float(part_2.split("<=====>")[1])
        time_i.append(this_timestamp)
        yaw_i.append(this_yaw)
    if i == len(logs_non_blank_line) - 1:
        print("all lines checked! save data...")
        list_time[current_test_index] = time_i
        list_yaw[current_test_index] = yaw_i
        current_test_index += 1
        time_i = []
        yaw_i = []
        print("save data finished!")

print("file 10-17-continuous cu.txt process finished!")
print(f"random target seq is {list_target_angle}")

continuous_cue_list_target_angle = list_target_angle
continuous_cue_list_time = list_time
continuous_cue_list_yaw = list_yaw

# #############################
with open(r"C:\Users\jiant\Desktop\data\turn-to-target\10-17-once verbal.txt", 'r', encoding='UTF-8') as f:
    logs = f.read().split('\n')
    logs_non_blank_line = [i for i in logs if i != '']
print("start processing file is : 10-17-once verbal.txt")
tmp_list = [i for i in logs_non_blank_line if i.__contains__("************ CoinSpin.cs, turn-in-space exp log : Coin init at random angle:")]
print(f"turn to target exp num is : {len(tmp_list)}")
print(f"non-blank line count is : {len(logs_non_blank_line)}")

list_target_angle = [None for i in tmp_list]  # 数组长度为实验次数
list_time = [None for i in tmp_list]  # 数组长度为实验次数
list_yaw = [None for i in tmp_list]  # 数组长度为实验次数

time_i = []  # 数组长度为 具体某次转向实验 的数据长度，每次都可能会不一样
yaw_i = []  # 数组长度为 具体某次转向实验 的数据长度，每次都可能会不一样

current_test_index = -1
for i in range(len(logs_non_blank_line)):
    if logs_non_blank_line[i] == "=======this is a gap=======":
        # 存起来上一次的数据
        if current_test_index != -1:
            list_time[current_test_index] = time_i
            list_yaw[current_test_index] = yaw_i
        # index 增加，并清空临时数组放下一轮的数据
        current_test_index += 1
        time_i = []
        yaw_i = []
        print(f"processing {current_test_index} th test data... ")
        continue
    if logs_non_blank_line[i].__contains__("************ CoinSpin.cs, turn-in-space exp log : Coin init at random angle:"):
        list_target_angle[current_test_index] = logs_non_blank_line[i].split("random angle:")[1]
        continue
    if logs_non_blank_line[i].__contains__("current yaw from IMU is ......."):
        part_2 = logs_non_blank_line[i].split(".......")[1]
        this_timestamp = int(part_2.split("<=====>")[0])
        this_yaw = float(part_2.split("<=====>")[1])
        time_i.append(this_timestamp)
        yaw_i.append(this_yaw)
    if i == len(logs_non_blank_line) - 1:
        print("all lines checked! save data...")
        list_time[current_test_index] = time_i
        list_yaw[current_test_index] = yaw_i
        current_test_index += 1
        time_i = []
        yaw_i = []
        print("save data finished!")

print("file 10-17-once verbal.txt process finished!")
print(f"random target seq is {list_target_angle}")

once_cue_list_target_angle = list_target_angle
once_cue_list_time = list_time
once_cue_list_yaw = list_yaw

print("done for all log data process, ready to fly...")


def plot_by_angle(angle):
    str_angle = str(angle)

    continuous_verbal_indexes = [i for i, x in enumerate(continuous_verbal_list_target_angle) if x == str_angle]
    cv_timestamp_a = []
    cv_yaw_a = []
    for index in continuous_verbal_indexes:
        cv_timestamp_a.append(continuous_verbal_list_time[index])
        cv_yaw_a.append(continuous_verbal_list_yaw[index])

    continuous_verbal_indexes = [i for i, x in enumerate(continuous_cue_list_target_angle) if x == str_angle]
    cc_timestamp_a = []
    cc_yaw_a = []
    for index in continuous_verbal_indexes:
        cc_timestamp_a.append(continuous_cue_list_time[index])
        cc_yaw_a.append(continuous_cue_list_yaw[index])

    continuous_verbal_indexes = [i for i, x in enumerate(once_cue_list_target_angle) if x == str_angle]
    ov_timestamp_a = []
    ov_yaw_a = []
    for index in continuous_verbal_indexes:
        ov_timestamp_a.append(once_cue_list_time[index])
        ov_yaw_a.append(once_cue_list_yaw[index])

    for j in range(len(cv_timestamp_a)):
        # timestamp to seconds
        start_timestamp = cv_timestamp_a[j][1]
        # tmp_timestamp = [i - start_timestamp for i in cv_timestamp_a[j] if 1]  # 变成毫秒
        tmp_timestamp = [(i - start_timestamp) / 1000 for i in cv_timestamp_a[j] if 1]  # 变成秒
        cv_timestamp_a[j] = tmp_timestamp
        # yaw to yaw diff
        tmp_yaw = [float(angle) - i for i in cv_yaw_a[j] if 1]
        cv_yaw_a[j] = tmp_yaw
        # 去掉第一个，因为第一个有启动等待时间
        del (cv_timestamp_a[j][0])
        del (cv_yaw_a[j][0])

    for j in range(len(cc_timestamp_a)):
        start_timestamp = cc_timestamp_a[j][1]
        tmp_timestamp = [(i - start_timestamp) / 1000 for i in cc_timestamp_a[j] if 1]
        cc_timestamp_a[j] = tmp_timestamp

        tmp_yaw = [float(angle) - i for i in cc_yaw_a[j] if 1]
        cc_yaw_a[j] = tmp_yaw
        del (cc_timestamp_a[j][0])
        del (cc_yaw_a[j][0])

    for j in range(len(ov_timestamp_a)):
        start_timestamp = ov_timestamp_a[j][1]
        tmp_timestamp = [(i - start_timestamp) / 1000 for i in ov_timestamp_a[j] if 1]
        ov_timestamp_a[j] = tmp_timestamp

        tmp_yaw = [float(angle) - i for i in ov_yaw_a[j] if 1]
        ov_yaw_a[j] = tmp_yaw
        del (ov_timestamp_a[j][0])
        del (ov_yaw_a[j][0])

    # 创建figure窗口，figsize设置窗口的大小
    plt.rcParams['figure.figsize'] = (10, 10)  # 单位是inches
    # fig, axs = plt.subplots(2, 1)
    # x y 标注
    plt.ylabel('angle difference (°)', fontdict={'family': 'Calibri', 'size': 25})
    plt.xlabel('time (s)', fontdict={'family': 'Calibri', 'size': 25})
    plt.title(f"target angle {angle}°", fontdict={'family': 'Calibri', 'size': 25})
    # 设置坐标轴范围
    # plt.xlim((1, 5.7))
    # plt.ylim((40, 90))
    # 坐标轴取点间隔
    # my_x_ticks = np.arange(0, 6, 1)
    # my_y_ticks = np.arange(0.5, 0.9, 0.1)
    # 坐标轴字体

    plt.xticks(fontproperties='Calibri', size=25)
    plt.yticks(fontproperties='Calibri', size=25)
    # plt.xticks([1.3, 2.3, 3.3, 4.3, 5.3], [r'$n$', r'$s$', r'$m$', r'$l$', r'$x$'], size=25)

    # 画出线图
    for j in range(len(cv_timestamp_a)):
        x_time = [i for i in cv_timestamp_a[j] if i <= 10]
        y_yaw_diff = [cv_yaw_a[j][i] for i in range(len(x_time))]
        plt.plot(x_time, y_yaw_diff, color='#d62728')  # 橙色 ff7f0e
    for j in range(len(cc_timestamp_a)):
        plt.plot(cc_timestamp_a[j], cc_yaw_a[j], color='#2ca02c')  # 绿色 2ca02c
    for j in range(len(ov_timestamp_a)):
        plt.plot(ov_timestamp_a[j], ov_yaw_a[j], color='#1f77b4')  # 蓝色 1f77b4

    plt.hlines(2.5, 0, 10, colors="r", linestyles="dashed")
    plt.hlines(-2.5, 0, 10, colors="r", linestyles="dashed")

    plt.legend(prop={'family': 'Calibri', 'size': 25})
    plt.show()
```
